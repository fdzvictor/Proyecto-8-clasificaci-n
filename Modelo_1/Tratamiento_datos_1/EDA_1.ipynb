## Estructura del Proyecto

Este proyecto est√° dividido en varias etapas las cuales son: 

1. **Exploraci√≥n de los datos**: Debes analizar las caracter√≠sticas disponibles y tomar las primeras decisiones clave:

3. **Preparaci√≥n del Modelo**: Antes de construir el modelo, debes preparar los datos adecuadamente. Este paso incluye:

        - etc. 

    - **Optimizaci√≥n del modelo**: Ajusta hiperpar√°metros para mejorar el desempe√±o, utilizando herramientas como GridSearchCV.


5. **Conclusiones**: La etapa final consiste en analizar los resultados obtenidos y proponer estrategias pr√°cticas. Esto incluye:

# EDA

Antes de empezar a tratar los datos, uniremos todos los dataframes por una key com√∫n ("EmployeeID") para poder trabajar m√°s f√°cilmente

````python
df_a = df_survey.merge(df_manager, on= "EmployeeID", how = "inner")
df = df_a.merge(df_general, on= "EmployeeID", how = "inner")
df
````

Comenzamos con una visualizaci√≥n r√°pida de los datos, que nos aporta informaci√≥n preliminar. En primer lugar podemos observar que Attrition (Nuestra variable objetivo) Es dicot√≥mica y est√° desbalanceada. Existen algunas columnas con valores √∫nicos tambi√©n. Resulta intrigante la bajada de empleados a partir de los 10 a√±os en la compa√±√≠a. Igualmente hay poca gente que haya trabajado m√°s de 10 a√±os.

````python
visualizar = s_eda.Visualizador(df)
visualizar.plot_numericas(tamano_grafica=[25,8])
df_num,df_cat =visualizar.separar_dataframes()
````

En la visualizaci√≥n de las categ√≥ricas, se aprecia que la mayor√≠a de empleados no suelen viajar por motivos de negocios. Suelen trabajar tambi√©n en R&D y tienen carreras en ciencias de la salud. Hay mayor√≠a de hombres

A simple vista hay tres columnas con valores √∫nicos (Standard Hours, EmployeeeCount y Over18). Como no aportan, las dropeamos

````python
df = df.drop(columns = ["Over18","EmployeeCount","StandardHours"])
````

Convertiremos tambi√©n nuestra variable objetivo a dicot√≥mica. S√≠ = 1, No = 0

````python
df["Attrition"] = df["Attrition"].map({"Yes": 1, "No": 0})
````

Hemos creado un dataframe donde poder ver las relaciones entra variables. Resultan especialmente interesantes aquellas entre "PerformanceRating" y "PercentSalaryHike", ", "Age" y "TotalWorkingYears". "YearsWithCurrManager" y "YearsAtCompany". "YearsSinceLastPromotion" y "Yearswithcurrentmanager". No encontramos ninguna relaci√≥n fuerte con "Attrition"

````python
visual = s_eda.Visualizador(df)
visual.dataframe_correlaci√≥n()
````

Con motivo de evitar la autocorrelaci√≥n en el modelo, eliminaremos "PercentSalaryHike", Pues PerformanceRating me parece un indicador m√°s f√°cil de entender y ordenar. De igual forma eliminaremos "YearsAtCompany", pues es una variable altamente correlacionada con otras y podemos usar "Age" para sustituirlas. Eliminaremos "YearsSinceLastPromotion" para evitar autocorrelaci√≥n con "YearsWithCurrManager". Esta √∫ltima tiene mayor correlaci√≥n con la variable objetivo, por tanto la eliminaremos. Por √∫ltimo eliminaremos tambi√©n "Age" frente a "TotalWorkingyears" por el mismo motivo

````python
df.drop(columns = ["PercentSalaryHike","YearsAtCompany","YearsSinceLastPromotion","Age"],inplace = True)
````

En el excel se indica que el salario est√° en Rupias. Lo pasamos a d√≥lares para tener una m√©trica m√°s f√°cil de interpretar (A d√≠a 29/11/2024 el cambio es 1 Rupia = 0.012$)
