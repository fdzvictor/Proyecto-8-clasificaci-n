Imputamos los nulos de la columna por iterative imputer y los redondeamos 

```python
tratar_nulos = s_eda.ImputarNulos(df)
df_imputado = tratar_nulos.imputar_iterative(lista_columnas= ["EmployeeID","EnvironmentSatisfaction","JobSatisfaction","WorkLifeBalance","NumCompaniesWorked"])
df_imputado = df_imputado.round()

df_imputado
```

Pasamos todas las columnas a "int" por motivos de limpieza y para poder concatenar el dataframe imputado con el dataframe original

```python
df_imputado = df_imputado[["EmployeeID_iterative","EnvironmentSatisfaction_iterative","JobSatisfaction_iterative","WorkLifeBalance_iterative","NumCompaniesWorked_iterative"]].astype("int")
```

Creamos un df con los nulos ya imputados y mantenemos las columnas imputadas como unicas. Por último comprobamos que efectivamente no hay más nulos.

```python
df_nonulos = df.join(df_imputado, lsuffix="EmployeeID", rsuffix="EmployeeID_iterative")
df_nonulos.drop(columns = ["EmployeeID","EnvironmentSatisfaction","JobSatisfaction","WorkLifeBalance","NumCompaniesWorked"],inplace = True)
df_nonulos.isna().sum()
```

Por último convertimos las variables que son categorías en el dicconario como categorías

```python
df_nonulos[["Education",
"EnvironmentSatisfaction_iterative",
"Gender",
"JobSatisfaction_iterative",
"MaritalStatus",
"PercentSalaryHike",
"WorkLifeBalance_iterative"]].astype("category")
```
